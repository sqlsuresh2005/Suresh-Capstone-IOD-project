```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Predict Students Dropout.csv')

# Display the shape of the dataset
print("Dataset Shape:", data.shape)

# Display data types and non-null counts
print("\nData Info:")
print(data.info())

# Summary statistics for numerical features
print("\nSummary Statistics for Numerical Features:")
print(data.describe())

# Summary statistics for categorical features
print("\nSummary Statistics for Categorical Features:")
print(data.describe(include='object'))

# Check for missing values
missing_values = data.isnull().sum()
print("\nMissing Values:")
print(missing_values[missing_values > 0])

# Plot histograms for numerical features
data.hist(bins=30, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Features')
plt.show()

# Bar plot for marital status distribution
plt.figure(figsize=(10, 6))
data['marital_status'].value_counts().plot(kind='bar')
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Correlation matrix for numerical features
correlation_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Dropout rates by course enrolled
dropout_rates = data.groupby('course_enrolled')['status'].value_counts(normalize=True).unstack()
dropout_rates.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Dropout Rates by Course Enrolled')
plt.xlabel('Course Enrolled')
plt.ylabel('Proportion')
plt.legend(title='Status', loc='upper right')
plt.show()

# Box plot for age to identify outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['age'])
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()

# Conclusion: Summarize findings
print("\nConclusion:")
print("1. Analyze trends in dropout rates by course and marital status.")
print("2. Investigate correlations between parental education and student success.")
print("3. Address any missing values or outliers identified.")
```
